```python
data = pd.read_csv('data_salaries_india.csv')
data
```

Converting Salaries into an intereger type, changing column name instead

```python
data['Salary'] = data['Salary'].str.replace('â‚¹', '')
data['Salary'] = data['Salary'].str.replace('/yr', '')
data['Salary'] = data['Salary'].str.replace(',', '')
data['Salary'] = data['Salary'].astype(int)
data['Yearly Salary, â‚¹'] = data['Salary']
data = data.drop('Salary', axis=1)
data
```

Let's sort and see if there are any Job Titles that can be combined into same one

```python
jobs = data['Job Title'].sort_values()
jobs.unique()
```

```python
data['Job Title'].replace('Lead Data Scientist', 'Data Science Lead', inplace=True)
data['Job Title'].replace('Data Science', 'Data Scientist', inplace=True)
data['Job Title'].replace('Machine Learning Data Associate I', 'Machine Learning Associate', inplace=True)
data['Job Title'].replace('Machine Learning Data Associate II', 'Machine Learning Associate', inplace=True)
data['Job Title'].replace('Associate Machine Learning Engineer', 'Machine Learning Associate', inplace=True)
data['Job Title'].replace('Machine Learning Data Associate', 'Machine Learning Associate', inplace=True)
data['Job Title'].replace('Machine Learning Software Engineer', 'Machine Learning Engineer', inplace=True)
data['Job Title'].replace('Software Engineer - Machine Learning', 'Machine Learning Engineer', inplace=True)

jobs = data['Job Title'].sort_values()
jobs.unique()
```

```python
data['Job Title'].value_counts()
```

Based on the amount of total data rows, it is safe to say that the low (below 10) positions are so few and are not represenatative of the general data. therefore, they will be removed from the data.

As is to happens, those roles include titles like Consultant and Trainee, where the salary data would not be representative, as stated above. Same applies for roles like Director and CEO, where those positions have significantly larger salary values.

```python
dropjobs = data['Job Title'].value_counts()
dropjobs_df = dropjobs.reset_index()
dropjobs_df.columns = ['name', 'count']
dropjobs = dropjobs_df[dropjobs_df['count'] >= 10]
mask = data['Job Title'].isin(dropjobs['name'])
data = data[mask]
```

```python
data['Job Title'].value_counts()
```

```python
data['role_fact'], unique_names = pd.factorize(data['Role'])
data
```

Now, we can do a correlation

```python
data.corr(method="spearman")
```

Based on this, we can say that having a "Manager" role usually leads to a higher salary
